## Analysis of Delta Hedging Project Results

### Plotting 100 of the 100 sample paths

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
stocks_path = pd.read_csv('stocks.csv', usecols = range(100), header = None)
stocks_path
```

```python
stock_sample = stocks_path.sample(n = 100, random_state = 1) #Randomly Sample 100 out of 1000
for i in stock_sample.index:
    plt.plot(stocks_path.iloc[i])

plt.title('Simulations Plot of 100 Sample Paths')
plt.xlabel('Steps')
plt.ylabel('Stock Price')
plt.show()
```

```python
option_path = pd.read_csv('options.csv', usecols = range(100), header = None)
option_path
```

```python
stock_sample = option_path.sample(n = 100, random_state = 1) #Randomly Sample 100 out of 1000
for i in option_path.index:
    plt.plot(option_path.iloc[i])

plt.title('Simulations Plot of 100 Sample Paths')
plt.xlabel('Steps')
plt.ylabel('Option Price')
plt.show()
```

```python
hedging_error = pd.read_csv('hedge_error.csv', 
                            usecols = range(99), 
                            header = None)
hedging_error
```

```python
# stock_sample = stocks_path.sample(n = 100, random_state = 1) #Randomly Sample 100 out of 1000
for i in hedging_error.index:
    plt.plot(hedging_error.iloc[i])

plt.title('Plot of 1000 delta hedging portfolios')
plt.xlabel('Steps')
plt.ylabel('Hedging Error')
plt.show()
```

```python
plt.hist(hedging_error.values.ravel('F'), bins=30)
plt.title('Distribution of Cumulative Hedging Errors')
plt.xlabel('Hedging Error')
plt.ylabel('Count')
plt.show()
```
